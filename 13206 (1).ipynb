```python
import nltk
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re

# Ensure required NLTK resources are downloaded
nltk.download('punkt')
nltk.download('stopwords')

# Sample text for the word cloud
text = """
Tokenization is the first step in text analytics. The process of breaking down a text 
paragraph into smaller chunks such as words or sentences is called Tokenization. 
Text analytics is used to analyze large amounts of textual data. Natural language processing 
is a crucial component of text analytics that helps identify patterns, sentiment, and meaning.
"""

# Clean the text (optional: remove non-alphabetic characters)
text = re.sub('[^a-zA-Z]', ' ', text)

# Tokenization
tokens = word_tokenize(text.lower())

# Remove stop words
stop_words = set(stopwords.words("english"))
filtered_tokens = [word for word in tokens if word not in stop_words]

# Create a word cloud
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(" ".join(filtered_tokens))

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")  # No axes for the word cloud
plt.show()
```
